Let's say that we're accurately measuring the temperature in both Madison and Minneapolis, but then our temperature sensor in Minneapolis breaks. We could easily install a new sensor, but we would prefer to estimate the temperature in Minneapolis based on the temperature in Madison. This process of estimation has broader implications, including (say) predicting one stock price based on another. Obviously, this problem is much more complex.

We're collecting data accurately. So, let's gather the data. In this, we're estimating $X$ from $Y$.

```python
y = read_temp('madison_old.csv')
x = read_temp('minneapolis_old.csv')

difference = abs(x - y)
print "Mean difference: ", mean(difference)

# plotting the difference between the two cities
figure()
plot(difference, 'o')
title('\\textrm{Temperature difference}')
ylabel('$^\circ$\\textrm{F}')
xlabel('\\textrm{Day}')
savefig('temp_diff.png', dpi=300)
show()
```

$\newcommand{\el}{l}$
$\newcommand{\ex}{\mathbb{E}}$



This linear estimation is rather simple. We're only saying that the temperature in Minneapolis is (say) 80% of Madison's temperature plus some constant. But, this simple estimation only requires simple statistics: the mean, variance and cross-correlation. Given the data we've collected in the past, these are simple to collect.

```python
Ex = mean(x)
Ey = mean(y)
Ey2 = var(y) + Ey**2
Ex2 = var(x) + Ex**2

Cov_xy = cov(x,y)[0,1]
Exy = Cov_xy + Ex * Ey
```

But now, let's predict the temperature. We want to minimize the error, or in math language, we want to minimize the $\el_2$ error: $ \min \ex (X - (aY+b)^2 $

Then following the math,

$$
\min \ex (X - (aY+b)^2) = \min -2\alpha \ex(XY) - 2\beta \ex(X)+\ex(X^2)+\alpha^2\ex(X^2)+\beta^2+2\alpha\beta\ex(X)
$$

This function is concave, and we can minimize it by taking two derivatives and setting them to 0. 

$$\frac{d}{d\alpha} = 0 = -2 \ex(XY) + 2 \alpha\ex(Y^2) + 2\beta \ex(y) $$

$$\frac{d}{d\beta} = 0 = -2 \ex(X) + 2\beta + 2\alpha\ex(X) $$

This equation is linear, meaning we can solve it using linear algebra:

```python
A = array([[2*Ey2, 2*Ey], [2*Ex, 2]])
b = array([2*Exy, 2*Ex])
c = solve(A, b)

a = c[0]
b = c[1]
```

We have found our $\alpha$ and $\beta$. But how good is our prediction?

```python
y_new = read_temp('madison_new.csv')
x_new = read_temp('minneapolis_new.csv')

x_predict = a*y_new + b

error = abs(x_predict - x_new)
print "Mean error: ", mean(error)

linear = linspace(0, 365, num=365)
o = ones(365)

figure()
plot(x_new, 'b.')
plot(y_new, 'g.')
xlabel('Day')
ylabel('Temperature')
savefig('pred_temp.png', dpi=300)
show()

figure()
plot(error, '.', label='Predicted difference')
plot(abs(x-y), '.', label='Original difference')
#plot(linear, o*mean(error), 'r-')
ylabel('\\textrm{Temperture difference}')
xlabel('\\textrm{Day}')
legend()
savefig('pred_diff.png', dpi=300)
show()
```

This is *exactly* what we hope for. Because these two locations are close, this is a good estimator. The error is nearly the same as the difference in the original (or training) data.
