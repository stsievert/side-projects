### Python
Python, by itself, is *not* a good scietific programming language. It is dreadfully slow and based mostly on `for` loops -- not much is vectorized. The main libraries and packages that make it good are [NumPy][np], [SciPy][sp] and [matplotlib][plt] (as well as many others). The base library, NumPy, contains fast routines for common mathematic and scientific operations. SciPy is another set of more advanced routines that depend on NumPy, and matplotlib is for visualizing your data. It should also be mentioned that [Pandas][pd] and [scikit-learn][sk] are good for input/output and writing really advanced algorithms (academics: that means you). 

To possibly speed up your (already optimized!) code, [Numba][nb] is just-budding JIT and [Cython][cy] is a long-existing C interface. When your code is already all-vectorized, Numba doesn't help much, but if you even have one for-loop, Numba helps. But be warned: it's very new, hence it doesn't work with all the code yet. If your code runs normally, but throws errors under Numba, don't use Numba.

This is just the surface of the packages available. To get the whole scientific stack, look at [Anaconda][ana] or the [Enthought Python distro][enthought] for full Python distros you can modify, or, for an IDE, look at [Spyder][spy].

### The basics
Python is not like Matlab. With Matlab, there are weird things such as that you can't have a script and function definitions in the same file. With Python, it's just as if you running everything at the prompt, meaning you can have declarations and use of those declarations in the same file.

#### Booleans
Besides `True` and `False`, the biggest thing here is `None`. If you have `return None`, that means that nothing is returned; it just ends the definition.

#### Importing functions
Let's say we have two files, called `script.py` and `functions.py`. `script.py`, as the name suggests, is just a script that uses functions in `functions.py`. To use these functions, we can write 

* `from functions import *`-- imports all the functions in the global namespace
* `import functions as f` -- means we can do `some_value = f.some_function_name(...)`
* `from functions import some_function_name` -- only import `some_functions_name`.

#### Docstrings
When you define a function, a doc-string shows up in other interfaces to describe the function. The begin and end with three `"` right below the declaration.



```python
def some_function(x):
    """ A function that doesn't do much """
    return x**2
```

#### Your interpetter
Use the [IPython][ipy] interpetter. It gives you tab-completion, history-completion, doc-strings, magic shell commands, all that stuff. The biggest one I use is `%run <filename>`. That puts all of the variables in your namespace, meaning you can see stuff inside your script.

When I'm writing a function, I don't include it in a `def ...:` at first. Instead, write my function just as a script. This means that I can use IPython's `%run` to ensure that my function is running right. After I've written it, I wrap it in a function.

[ipy]:http://ipython.org

Functions are objects like anything else. That means we can rename then or return them.

```python
def return_a_function(x):
    if x < 0:
        return np.dot
    if x > 0:
        return np.cross

that_function = return_a_function(-4);
print that_function

x = arange(16).reshape(4,4)
y = eye(4)
y[0, 1] = -1; # the first row, the second entry (indexing starts at 0)
z = that_function(x, y)
print z
```

### Packages
When you're first starting out, it can be annoying to write function names. For example, I could have written this when I was first starting:

Before we start any of the below, the designers of Python made a bad choice. They said that `int/int = int`, meaning `1/4 = 0` while `1.0/4 = 0.25`. Python 3 (we're at Python 2.7) has this feature. The made a module called `__future__`, meaning we can do

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100)
y = 1 - np.e**(-x/10)

plt.figure()
plt.plot(x, y, 'b.')
plt.show() # would not be shown if it wasn't for the next cell
# if division wasn't imported, 2/10 ==> 0, 9/10 ==> 0, etc. jumpy in the graph.
```

That's messy and icky and too many characters. Wouldn't it be nice if there was a package that (every?) function in NumPy and matplotlib.pyplot into your global namespace as well as importing `np` and `plt`? Well, it exists.

```python
from pylab import *

x = linspace(0, 100)
y = 1 - e**(-x/10)

figure()
plot(x, y, 'b.')
show()
```

#### Input and Output (Pandas)
Input and output can be a pain: parsing a CSV file by yourself isn't fun. That's where pandas comes in.

```python
from pandas import read_csv
from matplotlib import rc

# let's use latex in the title/labeling/everywhere
rc('text', usetex=True)

x = read_csv('image.csv', header=None)
print x

# making it an array
y = asarray(x)

# pylab doesn't replace the built in functions -- use "np." for that
print np.round(y, decimals=2)

imshow(x, cmap='gray')
colorbar()
axis('off')
title('\\textrm{Lena}')
# slash is the escape char -- \\ gives you a \, and that's interpetted by LaTeX
show()
```
