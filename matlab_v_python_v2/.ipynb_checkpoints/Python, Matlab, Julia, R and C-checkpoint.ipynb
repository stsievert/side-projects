Let's test the speeds of various programming languages. We'll do simple tests in Matlab, Python, R, Julia and C. We'll look at some simple approaches that are consistent across various languages.

```python
from time import time

def plainFor():
    ans = 0
    for i in arange(1e6):
        ans += i
def vecFor():
    x = arange(1e8)
    y = sum(x)

def svdTime():
    n = 524
    x = rand(n,n)
    u,s,v = svd(x)

def cumSumTime():
    x = arange(1e7)
    y = cumsum(x)
    
def euler():
    a = arange(1,1e3)
    b = arange(1,1e3)
    A, B = meshgrid(a, b)
    c2 = A**2 + B**2;
    C = np.sqrt(c2)
    i = argwhere(A+B+C == 1000)
    ans = A[i[0], i[1]] * B[i[0], i[1]] * C[i[0], i[1]]
    
start = time(); plainFor(); end = time()
numpy_for = end - start

start = time(); vecFor(); end = time()
numpy_vecfor = end - start

start = time(); svdTime(); end = time()
numpy_svd = end - start

start = time(); cumSumTime(); end = time()
numpy_cumsum = end - start

start = time(); euler(); end = time()
numpy_euler = end - start

numpy = array([numpy_for, numpy_vecfor, numpy_svd, numpy_cumsum])
```
